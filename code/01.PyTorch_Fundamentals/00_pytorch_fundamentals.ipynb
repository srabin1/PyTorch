### Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning

1. Element-wise multiplication
2. Matrix multiplication (dot product)

There are two main rules that performing matrix multiplication needs to satisfie:

1. The **inner dimensions** must match:
* `(3,2) @ (3,2)` won't work
* `(2,3) @ (3,2)` will work
* `(3,2) @ (2,3)` will work

2. The resulting matrix has the shape of the **outer dimensions**:
* `(2,3) @ (3,2) -> (2,2)`
* `(3,2) @ (2,3) -> (3,3)`


```python
# Element wise multiplication
print(tensor, "*", tensor)
print("Equals: {}".format(tensor * tensor))
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

what we are doing in tensor multiplication:
resource: https://matrixmultiplication.xyz

```python
# shapes for matrix multiplication
# torch.matmul == torch.mm
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])
tensor_B = torch.tensor([[7,10],
                        [8,11],
                        [9,12]])

torch.mm(tensor_A, tensor_B)
```

```python
tensor_A.shape, tensor_B.shape
```

To fix our tensor's shape issues, we can manipulate the shape of one of our tensors using a **transpose**

A **transpose** switches the axes or dimensions of a given tensor.

```python
tensor_B.T
```

```python
tensor_B
```

```python
torch.matmul(tensor_A, tensor_B.T)
```

```python
# The matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}")
print (f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("output: \n")
output = torch.mm(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

## PyTorch tensors and NumPy

NumPy is a popular scientific Python numerical computing library.

And because of this, PyTorch has functionality to interact with it.

* Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor -> NumPy -> `torch.Tensor.numpy()`

The default data type of NumPy is float64 while the default data type of tensor is float32, that causes error, so you need to make them consistent

```python
# NumPy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) # warning: when converting from numpy -> pytorch, pytorch reflects numpy's default datatype of float64 unless specified otherwise
array, tensor
```

```python
array.dtype
```

```python
tensor.dtype
```

```python
torch.arange(1.0,8.0).dtype
```

If you want to change the data type to float 32:

* array = np.arange(1.0, 8.0)

* tensor = torch.from_numpy(array).type(torch.float32)

```python
# Change the value of the array, what will this do to `tensor`? No

array = array + 1
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones (7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# change the tensor, what happens to `numpy_tensor`? it remains unchanged
tensor = tensor +1
tensor, numpy_tensor
```

## Reproducibility (trying to take random out of random)

In short how a neural network learns:

`start with random numbers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> again...

To reduce the randomness in neural networks and PyTorch comes the concept of a **random seed**.

Essentially what the random does is "flavour"the randomness.

```python
torch.rand(3,3)
```

```python
import torch
# create two random tensors
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A==random_tensor_B)
```

```python
# Let's make some random but reproducible tensors
import torch
# set the random seed
RANDOM_SEED = 42

# we should use torch.manual-seed() everytime before creating a new tensor
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C==random_tensor_D)
```
